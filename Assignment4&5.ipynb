```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset("titanic")
```

```python
print(dataset.head())
```

```python
sns.set(style="whitegrid")
```

```python
mean = dataset.mean()
max = dataset.max()
min = dataset.min()
print("Get mean of entire DataFrame:\n", mean)
print("Get max of entire DataFrame:\n", max)
print("Get min of entire DataFrame:\n", min)
```

```python
sns.countplot(x='survived', data=dataset)
plt.title('Count of Survivors and Non-Survivors')
plt.show()
```

```python
sns.countplot(x='survived', hue='sex', data=dataset)
plt.title('Survivors and Non-Survivors by Gender')
plt.show()
```

```python
sns.countplot(x='class', data=dataset)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
sns.histplot(dataset['age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=dataset)
plt.title('Survivors and Non-Survivors by Passenger Class')
plt.show()
```

```python
sns.histplot(x='fare', hue='survived', data=dataset, bins=30, kde=True)
plt.title('Distribution of Fares for Survivors and Non-Survivors')
plt.show()
```

```python
sns.countplot(x='survived', hue='sibsp', data=dataset)
plt.title('Survivors and Non-Survivors by Siblings/Spouses Aboard')
plt.show()
```

```python
dataset['title'] = dataset['who'] + ' ' + dataset['sex']
sns.countplot(y='title', data=dataset, order=dataset['title'].value_counts().index)
plt.title('Distribution of Passenger Titles')
plt.show()
```

```python
bins = [0,100,200,300,400,500]
sns.histplot(x='fare', data=dataset, bins=bins)
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='survived', y='age', hue='pclass', data=dataset, split=True)
plt.title('Violinplot of Age by Survival Status and Passenger Class')
plt.show()

sns.swarmplot(x='sex', y='age', hue='survived', data=dataset)
plt.title('Swarmplot of Age by Gender and Survival Status')
plt.show()

corr_matrix = dataset.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=dataset)
plt.title('Boxplot of Age by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
sns.boxplot(x='age', y='sex', hue='survived', data=dataset, orient='h')
plt.title('Horizontal Boxplot of Age by Gender and Survival Status')
plt.xlabel('Age')
plt.ylabel('Gender')
plt.show()
```

```python
numerical_columns = dataset.select_dtypes(include=['int64', 'float64']).columns

sns.set(style="ticks")
sns.pairplot(dataset, hue='survived', vars=numerical_columns)
plt.suptitle('Scatter Plots for Numerical Columns by Survival Status')
plt.show()
```
